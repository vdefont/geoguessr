```python
import numpy as np
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt
```

```python
w = 75
std = 12
m = np.zeros((w,w))
m[w//2,w//2] = 1
gf = gaussian_filter(m, std)
```

```python
W = 2.5
H = 2.75
back = np.zeros((int(w*H), int(w*W)))
coords_to_add={
    (0,0): 1.,
    (1.2,0.7): 1.,
    (1.4,0.4): 1.,
    (0.4,1.2): 0.5,
    (0.6,0.9): 0.5,
}
for (x, y), intensity in coords_to_add.items():
    back[int(y*w):int((y+1)*w),int(x*w):int((x+1)*w)] += gf * intensity
```

```python
plt.imshow(back, cmap=plt.cm.viridis, alpha=0.8)
plt.axis('off')
plt.savefig("gaussian.png", transparent=True)
plt.show()
```
