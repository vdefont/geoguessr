```python
import pandas as pd
```

```python
data = pd.read_csv("data.csv", index_col=0)
country_counts = data.country.value_counts().to_frame(name="count").reset_index()
country_counts = country_counts.rename(columns={"index": "country"})
```

```python
country_population = pd.read_csv("country_population.csv")
```

```python
cdata = country_counts.merge(country_population, on="country", how="left")
```

```python
cdata["pop_per_cnt"] = (cdata.population / cdata["count"]) / 1_000
```

```python
country_name = data[['country','country_name']].drop_duplicates()
country_to_name = dict(zip(country_name.country, country_name.country_name))
country_to_name['tw'] = 'Taiwan'
country_to_name['ru'] = 'Russia'
```

```python
cdata.country = cdata.country.map(country_to_name)
```

```python
cdata.pop_per_cnt = cdata.pop_per_cnt.round(1)
```

```python
cdata.population = (cdata.population/1_000_000).round(2)
```

```python
cdata.columns = [
    "Country",
    "Dataset Count",
    "Population (millions)",
    "Pop per Datum (thousands)",
]
```
