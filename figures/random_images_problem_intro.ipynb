```python
import torch
import numpy as np
import pandas as pd
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# Randomly chosen
IMGS = [65_177, 67_842, 94_297, 11_387]
```

```python
data = pd.read_csv("data.csv", index_col=0)
```

```python
rands = data.iloc[IMGS] # Last one is in Hawaii
rands.head()
```

```python
files = rands.index.tolist()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20,6))
fig.subplots_adjust(wspace=0.02,hspace=0.02)
for file, ax in zip(files, axes):
    im = Image.open(f"random_imgs/{file}.jpg")
    ax.imshow(im)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)

fig.text(0.5, 0.07, "(Images randomly selected from our dataset)", ha='center', fontsize=16)
txt = (
    "From left to right: South Africa, Italy, Switzerland, Hawaii"
)
fig.text(0.5, 0.13, txt, ha='center', fontsize=16, rotation=180)
plt.show()
```
